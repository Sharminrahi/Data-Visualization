# total number of video game consoles sold by Nintendo, Microsoft and Sony.

```python
#Calculate the total number of units sold for each company
company_units_sold = df5.groupby("Company")["Units_sold"].sum()
company_units_sold
```

```python
# Sort the DataFrame by total units sold for each console for each company
sorted_df5 = df5.sort_values(by=["Company", "Units_sold", "Console"], ascending=[False, True, False])
sorted_df5
```

```python
# Calculate the total number of units sold by the three companies
console_units_sold = sorted_df5.groupby(["Company", "Console"])["Units_sold"].sum()
console_units_sold
```

```python
#define the colors for the inner ring and outer ring
outer_colors = {"Microsoft": "#5599CC", "Nintendo": "#EA6727", "Sony": "#83C143"}
inner_colors = ["#156EAF", "#DB2018", "#56B45B"]
```

```python
# Define the figure and subplot
fig, ax = plt.subplots(figsize = (8,8))

outer_consoles = []
outer_values = []
outer_colors_list = []

for company in sorted_df5["Company"].unique():
    company_sales = console_units_sold.loc[company].sort_values(ascending = False)
    outer_consoles += list(company_sales.index)
    outer_values += list(company_sales.values)
    outer_colors_list += [outer_colors[company] for _ in range(len(company_sales))]

outer = ax.pie(
    outer_values,
    radius = 1,
    colors = outer_colors_list,
    labels = outer_consoles,
    counterclock = False,
    wedgeprops = dict(edgecolor = "w", linewidth = 1.5)
)

# Create the inner ring for all companies
inner = ax.pie(
    company_units_sold,
    radius = 0.6,
    colors = inner_colors,
    labels = company_units_sold.index,
    labeldistance = 0.5,
    wedgeprops = dict(edgecolor = "w", linewidth = 1)
)
ax.set_title("Best Selling Video Game Consoles")

plt.show()
```

```python
# Define the figure and subplot
fig, ax = plt.subplots(figsize = (8,8))


# Create the outer ring for all consoles
outer_consoles = []
outer_values = []
outer_colors_list = []

for company in sorted_df5["Company"].unique():
    company_sales = console_units_sold.loc[company].sort_values(ascending=False)
    outer_consoles += list(company_sales.index)
    outer_values += list(company_sales.values)
    outer_colors_list += [outer_colors[company] for _ in range(len(company_sales))]

# Replace label with empty string for the three consoles
outer_consoles[outer_consoles.index("Color TV-Game")] = ""
outer_consoles[outer_consoles.index("NES Classic")] = ""
outer_consoles[outer_consoles.index("SNES Classic")] = ""

outer = ax.pie(
    outer_values,
    radius = 1,
    colors = outer_colors_list,
    labels = outer_consoles,
    counterclock = False,
    wedgeprops = dict(edgecolor = "w", linewidth = 1.5)
)

# Create the inner ring for all companies
inner = ax.pie(
    company_units_sold,
    radius = 0.6,
    colors = inner_colors,
    labels = company_units_sold.index,
    labeldistance = 0.5,
    wedgeprops = dict(edgecolor = "w", linewidth = 1)
)
ax.set_title("Best Selling Video Game Consoles")

plt.show()
```

# Import venn diagram

```python
from matplotlib_venn import venn2
```

```python
df6 = pd.read_csv("olympic_athletes.csv", na_filter = False)
df6.head(10)
```

```python
df6.isnull().sum()
```

```python
df7 = df6[df6.Year.isin([2012, 2016])]
df7
```

```python
df7.Year.unique()
```

```python
df8 = df7.groupby(['Team', 'Year']).size().unstack()
df9 = df8.sort_values(by = 2012, ascending = False)
```

# Selected column for y axis

```python
df10 = df9.loc[["Great Britain", "United States", "Russia", "Australia", 
               "Germany", "China", "France", "Japan", "Italy", "Spain", "Brazil"]]
df10
```

## Convert the data to long format(melt)

```python
df10_long = pd.melt(df10.reset_index(), id_vars=['Team'], var_name='Year', value_name='Number of Athletes')
df10_long.head(15)
```

### The bar plot below shows the number of athletes that competed in the 2012 (dark blue) and 2016 (light blue) Olympics from a set of different countries. The Venn diagram acts as the legend – it informs the meaning of each colour.

```python
# Define a color palette
colors = plt.cm.tab20(range(2))

# Create the figure with two subplots
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (14, 6), 
                               gridspec_kw = {'width_ratios': [4, 1], 'height_ratios': [1.3]})

# Add the bar plot to the first subplot
sns.barplot(x = 'Number of Athletes', y = 'Team', data = df10_long, hue='Year', palette = colors, ax = ax1)

ax1.set(xlabel = 'Number of Athletes', xticks = [0, 100, 200, 300, 400, 500], ylabel = '', 
        title='Change in Number of Athletes per Country from 2012 to 2016')

ax1.legend_.remove()

# Add the venn diagram to the second subplot
venn = venn2(subsets = (2464, 2815, 1358), set_labels = ('2012', '2016'), set_colors = colors, alpha=0.8, ax = ax2)
venn.get_label_by_id('10')
venn.get_label_by_id('01')
venn.get_label_by_id('11')

# Adjust the margins to shift the venn diagram to the left
fig.subplots_adjust(left = 0.1, right = 0.9, bottom = 0.1, top = 0.9)

# Move the second subplot inside the first subplot
pos2 = ax2.get_position() # get the position of the second subplot

ax2.set_position([ax1.get_position().x0 + ax1.get_position().width - pos2.width, 
                  ax1.get_position().y0 + 0.08, pos2.width, pos2.height])

# Show the plot
plt.show()
```
