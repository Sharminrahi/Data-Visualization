The following violin plot shows the **height distributions of athletes** in the **Gymnastics, Cycling and Basketball** sports at the Olympic games between **2000 to 2016 (inclusive)**. The **line** inside each violin shows the location of the **median value.**

Recreate violin plot using Matplotlib. The three violins should be **purple, green and blue** in colour, including their filled colour and the colour of the median lines (Hint: consider the return values of the violin plot method). Include the x-axis and y-axis labels, as well as the plot title.

### Sports we want to include

```python
sports = ['Gymnastics', 'Cycling', 'Basketball']
```

### Filter the specified sports and time frame

```python
#df1 = df.loc[(df['Sport'].isin(sports)) & (df['Year'] >= 2000) & (df['Year'] <= 2016)]
```

```python
df1 = df[(df['Sport'].isin(sports)) & (df['Year'].between(2000, 2016))]
```

```python
df1
```

### Since some athletes may compete in multiple events or multiple Olympics, make sure to include only one instance of an athlete (no duplicates). Drop duplicates based on the athlete's ID

```python
df1 = df1.drop_duplicates(subset=['ID'])
df1
```

### Remove rows where the height is missing

```python
df1 = df1[~df1['Height'].isnull()]
df1
```

### New DataFrame with specific sports, year between 2000 to 2016 and heigth column

```python
df2 = df1[['ID', 'Sport', 'Height', 'Year']]
df2
```

### Now recreate violin plot using Matplotlib which represents the distributions of athletes in the Gymnastics, Cycling and Basketball sports at the Olympic games between 2000 to 2016 (inclusive).

```python
#Setup violin colours
colors = ['purple', 'green', 'blue']
```

```python
#setup figure and axes objects
fig, ax = plt.subplots(figsize = (8,6))

#Create the violins
violin = ax.violinplot([df2[df2['Sport']==sport]['Height'] for sport in sports],
                            showmedians = True,
                            showmeans = False,
                            showextrema = False,
                            vert = False,
                            points = 50,
                            widths = 0.8,
                            positions = [1,2,3])

bodies = violin['bodies']
medians = violin['cmedians']

for i, violin in enumerate(bodies):
    violin.set_facecolor(colors[i])
    violin.set_edgecolor('black')
    violin.set_alpha(0.35)
    medians.set_color(colors)
```

```python
#set the yaxis ticks and labels
ax.set_yticks([1,2,3])
ax.set_yticklabels(sports)

# set the x-axis limits and ticks
ax.set_xlim(120, 240)
ax.set_xticks([120, 140, 160, 180, 200, 220, 240])
xticklabels = [ '', '140', '160', '180', '200', '220', '' ]
ax.set_xticklabels(xticklabels, fontsize = 10) # set empty labels for 120 and 240


# add labels and title
ax.set_xlabel('Height (cm)')
ax.set_title('Height Distributions (2000 - 2016 Olympics)')

# show the plot
plt.show()
```

# Question 2
This line plot shows the **average height of male (blue) and female (red)** Olympians over the **history** of the **Summer** Olympics. The **shaded region** represents the **standard deviation about the average**. Recreate the above line plot using Matplotlib. Include the axes labels, plot title, and legend in the lower left corner. The average male heights should be blue and average female heights red. 

**Each athlete should only be represented once in a given year.** Each line should include shading that has width equal to twice the standard deviation, that is, the **top of the shaded region is the average value plus the standard deviation, and the bottom of the shaded region is the average value minus the standard deviation (hint: try fill_between()). The shading uses a transparency value of 0.1.**

### Filter only Summer olympics

```python
df3 = df[df.Season == "Summer"]
df3.head(10)
```

### Drop Duplicates

```python
df3 = df3.drop_duplicates(subset=['ID', 'Year'])
df3
```

### Remove the rows where heights is missing 

```python
df4 = df3[~df3['Height'].isnull()]
df4
```

## Group by Sex, Year and find average height of Male and Female

```python
df5 = df4.groupby(['Sex', 'Year'])['Height'].agg(['mean', 'std']).reset_index()
df5
```

### Extract the male and female data separately

```python
male_data = df5[df5['Sex'] == 'M']
female_data = df5[df5['Sex'] == 'F']
```

```python
print(male_data)
```

```python
print(female_data)
```

## Recreate the line plot using Matplotlib

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize = (8,6))

# plot the male and female lines with shading
ax.plot(male_data['Year'], male_data['mean'], color = 'blue', label = 'Male')

ax.fill_between(male_data['Year'],
                male_data['mean']- male_data['std'],
                male_data['mean'] + male_data['std'],
                color = 'blue',
                alpha = 0.1)

ax.plot(female_data['Year'], female_data['mean'], color = 'red', label = 'Female')

ax.fill_between(female_data['Year'],
               female_data['mean']- female_data['std'],
               female_data['mean'] + female_data['std'],
               color = 'red',
                alpha = 0.1);
```

```python
# Add labels, title, and legend
ax.set_xlabel('Year')
ax.set_ylim(140, 200)
ax.set_yticks([140, 150, 160, 170, 180, 190, 200])
yticklabels = ['140', '150', '160', '170', '180', '190', '200']
ax.set_yticklabels(yticklabels)
ax.set_ylabel('Height (cm)')
ax.set_title('Average Olympian Height (+/- std dev) in the Summer Olympics')
ax.legend(loc='lower left')

plt.show()
```

## Top 10 most decorated Olympians based on their total number of medal

### See more in Medal column

### Drop missing value of Medal

```python
df6 = df.dropna(subset = ['Medal'])
df6
```

```python
df.Name.value_counts()
```

### Groupby athletes name and count the number of Medals they won select top 20

```python
df7 = df6.groupby('Name')['Medal'].count().sort_values(ascending=False).head(20)
df7
```

```python
### Sort by medal count in descending order and select top 20
```

```python
#top_20 = df7.sort_values(ascending=False).head(20)
```

### Filter the original DataFrame to include only rows where the athlete name is in the top 20 list

```python
df8 = df6[df6['Name'].isin(df7.index)]
df8
```

```python
# Group by athlete name and medal type, and count the number of medals of each type
medal_counts = df8.groupby(['Name', 'Medal']).size().unstack(fill_value=0)
medal_counts
```

```python
# Calculate the total medal count for each athlete
medal_counts['Total'] = medal_counts.sum(axis=1)

# Sort by total medal count in descending order
medal_counts = medal_counts.sort_values(['Total', 'Gold'], ascending=[False, False])
top_10 = medal_counts.head(10)
```

```python
top_10
```

# Recreate Bar Plot

```python
# Create the bar plot for the top 10 athletes
fig, ax = plt.subplots()
colors = ['#FFD700', '#C0C0C0', '#B08D57']
top_10.loc[:, ['Gold', 'Silver', 'Bronze']].plot(kind='barh', stacked=True, color=colors, ax=ax)
ax.set_title('The Most Decorated Olympians of All Time')
ax.invert_yaxis()
ax.legend().set_visible(False)
ax.set_ylabel('')
plt.show()
```
